```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Military Expenditure.csv", index_col=1)
df.head()
```

```python
df.describe()
```

```python
df.isna().sum().plot(kind='bar', rot=90)
plt.show()
```

```python
df['Type'].value_counts(sort=True)
```

```python
df.sort_values(['1960', '1961']).head()
```

```python
df_countries = df[df['Type']=='Country']
df_countries = df_countries[df_countries['1960']!=0].sort_values('1960', ascending=False)
df_countries.head()
```

```python
df_countries[df_countries['1960']!=0].sort_values('2018', ascending=False).head()
```
