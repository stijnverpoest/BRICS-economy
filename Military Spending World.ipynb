```python
df_countriesmelt=df_countries.melt(id_vars=['Name', 'Type', 'Indicator Name'], var_name='year', value_name= 'budget')
df_countriesmelt = df_countriesmelt.sort_values(by=['Name', 'year'], ascending=True)
df_countriesmelt.head()
```

```python
df_countriesmelt.query('Name == "United States" and year=="2018"')
```

```python
df_countries_pivot=df_countriesmelt.pivot(index='year', columns='Name', values='budget')
df_countries_pivot.head()
```

```python
df_countries_pivot.plot()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
EUlist = ["Belgium", "Germany", "France", "Netherlands", "Luxembourg", "Ireland", "United Kingdom"]
df_countriesEU=df_countriesmelt[df_countriesmelt['Name'].isin(EUlist)]
```

```python
df_countriesEU_pivot = df_countriesEU.pivot(index='year', columns='Name', values='budget')
df_countriesEU_pivot.plot(title='Western Europe + Germany')
plt.show()
```

```python
def plot_benelux(a, b, axes, x, y, color, title):
    axes[a,b].plot(x,y, color=color)
    axes[a,b].set_xticks(axes[a,b].get_xticks()[::10])
    axes[a,b].set_title(title)
    
fig, ax = plt.subplots(2,2)
fig.tight_layout()
plot_benelux(0,0,ax,df_countriesEU_pivot.index, df_countriesEU_pivot['Netherlands'],'orange', "NL")
plot_benelux(0,1,ax,df_countriesEU_pivot.index, df_countriesEU_pivot['Belgium'], 'black', "BE")
plot_benelux(1,0,ax,df_countriesEU_pivot.index, df_countriesEU_pivot['Ireland'], 'green', "IR")
plot_benelux(1,1,ax,df_countriesEU_pivot.index, df_countriesEU_pivot['Luxembourg'], 'blue', "LU")
plt.show()
```

```python
def plot_timeseries(axes, x, y, marker, color, xlabel, ylabel):
    axes.plot(x,y, marker=marker, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.tick_params('y')
    axes.set_xticks(axes.get_xticks()[::10])
```

```python
df_countriesEU_pivot.loc[df_countriesEU_pivot['France'].idxmax()] #get the max value-date for france
```

```python
fig, ax = plt.subplots(1,1)
plot_timeseries(ax, df_countriesEU_pivot.index, df_countriesEU_pivot['France']/10, 'o', 'purple', 'year', 'spending (in Billions)')
ax.set_title("France military spending")
ax.annotate("66.88403 B",
            xy = ('2009', 6.688403e+9),
            xytext = ('2005', 2*10**9),
            arrowprops = {"arrowstyle":"->", "color":"gray"})
plt.show()
```
