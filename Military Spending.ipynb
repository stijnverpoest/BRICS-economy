```python
df = pd.read_csv("Military Expenditure.csv")
df.head()
```

```python
df_belgium = df[df.Name == "Belgium"]
df_belgium.head()
```

```python
df_belgium_years = df_belgium.iloc[:,4:]
df_belgium_years.head()
```

```python
df_belgium_years_rows = df_belgium_years.melt(var_name = "year", value_name = "expenses")
df_belgium_years_rows.head(15)
```

```python
tick_val= [0,1000000000,2000000000,3000000000,4000000000,5000000000,6000000000,7000000000] 
tick_lab=['0B','1B','2B','3B','4B','5B','6B','7B']

plt.scatter(x=df_belgium_years_rows.year, y=df_belgium_years_rows.expenses, c='green')
#plt.yscale('log')
plt.xlabel("Years")
plt.ylabel("Expenditure (USD)")
plt.title("Military Expenditure of Belgium from 1960-2018")
plt.yticks(tick_val, tick_lab)
plt.xticks(['1960','1970','1980','1990','2000','2010','2018'])
plt.text(35, 1000000000, "Conscription Suspended", fontsize='7',c='red') #35 --> 1995
plt.grid(True)
plt.show()
```
